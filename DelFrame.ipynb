```python
import numpy as np
import glob
import cv2
import os
```

```python
ii=0
for imagePath in glob.glob("EmojiImages" + "/*.png"):
    # extract the image filename (assumed to be unique) and
    # load the image, updating the images dictionary
    leftBorder = 0
    rightBorder = 159
    topBorder = 0
    downBorder = 159
    filename = imagePath[imagePath.rfind("/") + 1:]
    image = cv2.imread(imagePath)
    for i in range(0,159):
        isW = True
        for j in range(5,155):
            if image[i][j][0] < 250 or image[i][j][1] < 250 or image[i][j][2] < 250:
                isW = False
                break
        if isW:
            topBorder+=1
        else:
            break
            
    for i in range(0,159):
        isW = True
        for j in range(5,155):
            if image[159-i][j][0] < 250 or image[159-i][j][1] < 250 or image[159-i][j][2] < 250:
                isW = False
                break
        if isW:
            downBorder-=1
        else:
            break
            
    for j in range(0,159):
        isW = True
        for i in range(5,155):
            if image[i][j][0] < 250 or image[i][j][1] < 250 or image[i][j][2] < 250:
                isW = False
                break
        if isW:
            leftBorder+=1
        else:
            break
            
    for j in range(0,159):
        isW = True
        for i in range(5,155):
            if image[i][159-j][0] < 250 or image[i][159-j][1] < 250 or image[i][159-j][2] < 250:
                isW = False
                break
        if isW:
            rightBorder-=1
        else:
            break
    
    capacity = max(downBorder-topBorder+1,rightBorder-leftBorder+1)
    
    newIm = image[topBorder:(downBorder+1), leftBorder:(rightBorder+1)]
    
#     newOutIm = np.full((capacity,capacity,3),255)
    
#     if downBorder-topBorder < rightBorder-leftBorder:
#         bor = (capacity-downBorder+topBorder-1)//2
#         newOutIm[bor:(bor+downBorder-topBorder+1), 0:capacity] = newIm
#     else:
#         bor = (capacity-rightBorder+leftBorder-1)//2
#         newOutIm[0:capacity, bor:bor+rightBorder-leftBorder+1] = newIm
    cv2.imwrite("NewEmoji/"+filename, newIm)
#     print(filename+"  "+str(newIm.shape)+"  "+str(capacity))
#     if ii == 20:
#         break
#     ii+=1
```
